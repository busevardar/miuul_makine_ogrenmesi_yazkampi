```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


pd.set_option("display.max_columns",None)
pd.set_option("display.width",500)


df = pd.read_csv("datasets/breast_cancer.csv")
df = df.iloc[:,1:-1]
df.head()
```

```python
num_cols = [col for col in df.columns if df[col].dtypes in ["int","float"]]
num_cols
```

```python
corr = df[num_cols].corr()
corr
```

```python
# korelasyon iki veri arasındaki ilişkiyi ölçer. -1 ve +1 arasında değerler alır.
# -1 ve +1 e yakınlaştıkça ilişki kuvvetlenir denir.
# -1 : değişkenlerden biri artarken birinin azalması durumu
# +1 : değişkenlerden biri artarken diğerinin de artması durumu
# corr 0â yaklaştıkça ilişki azalıyor demektir.
```

```python
sns.set(rc = {'figure.figsize' : (12,12)})
sns.heatmap(corr, cmap = "RdBu")
plt.show()
```

```python
###########################
#Yüksek Korelasyonlu Değişkenlerin Sİlinmesi
###########################
```

```python
cor_matrix = df.corr().abs()
```

```python
upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k = 1).astype(np.bool))
#matrisi üst üçgen matris hale getirdik. Böylece satırda ve sütunda 
#aynı anda bulunup eşleşen elemanların sebep olduğu tekrarlamanın önüne geçtik.
upper_triangle_matrix
```

```python
#+ veya - 1 e aşırı yakınlaştığında ilişki maksimum olduğundan aslında aynı değişkenmiş gibi olurlar.
```

```python
drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > 0.90)]
drop_list
```

```python
cor_matrix[drop_list]
```

```python
df.drop(drop_list, axis = 1)
```

```python
def high_correlated_cols(dataframe, plot = False, corr_th = 0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),
                                                     k = 1).astype(np.bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc = {'figure.figsize' : (12,12)})
        sns.heatmap(corr, cmap = "RdBu")
        plt.show()
    return drop_list
```

```python
drop_list = high_correlated_cols(df)
drop_list 
```

```python
drop_list = high_correlated_cols(df, plot = True)
drop_list
```

```python
df.drop(drop_list, axis = 1)
```

```python
drop_list = high_correlated_cols(df.drop(drop_list, axis = 1), plot = True)
drop_list
```
